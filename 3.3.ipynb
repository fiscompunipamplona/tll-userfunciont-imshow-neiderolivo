```python
from math import sqrt, sin,pi
from numpy import empty 
from pylab import imshow, gray, show,hot

wavelength=5.0
k=2*pi/wavelength
xi0=1
separation=20
side=100
points=500
spacing=side/points
#calculando la posisicion de los centros de los circulos

x1=side/2 + separation/2
y1=side/2
x2=side/2 - separation/2
y2=side/2

#creando los vectores 
xi=empty([points,points],float)

#calculando los valores de los vectores 
for i in range (points):
    y=spacing*i
    for j in range (points):
        x=spacing*j
        r1=sqrt((x-x1)**2 + (y-y1)**2)
        r2=sqrt((x-x2)**2 + (y-y2)**2)
        xi[i,j]= xi0*sin(k*r1) + xi0*sin(k*r2)
    
#realizando el plot
imshow(xi,origin="lower",extent=[0,side,0,side])
hot()
show()
```

 En esta grafica se observa el fenómeno de interferencia debido a dos fuentes puntuales separadas a una cierta distancia una de la otra como se observaría en el experimento de las cubetas de ondas o en el experimento de Young, en esta se aprecia como se superponen las ondas creando varios frentes de ondas, unos que se destruyen y otros que se superponen, claramente se observan los máximos y mínimos de intensidad producidos por la superposición de las ondas.
